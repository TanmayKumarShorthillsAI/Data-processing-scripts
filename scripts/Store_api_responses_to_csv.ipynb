```python
import requests
import pandas as pd
```

```python
response = requests.get("https://datausa.io/api/data?drilldowns=Nation&measures=Population")
data = response.json()["data"]
data
```

```python
df = pd.DataFrame(data)
df
```

```python
def remove_duplicate_year(df, yr1, yr2):
    print(df[yr1].astype(int).equals(df[yr2].astype(int)))
    if df[yr1].astype(int).equals(df[yr2].astype(int)):
        df = df.drop([yr2], axis=1)
    return df
    # print(df)
```

```python
df
```

```python
nation = df["Slug Nation"].str.replace("-"," ")
nation
```

```python
def remove_duplicate_columns(df, col1, col2):
    if df[col1].str.lower().equals(df[col2].str.replace("-"," ")):
        df = df.drop(columns=[col2])
    return df
```

```python
df = remove_duplicate_year(df,"Year", "ID Year")
df = remove_duplicate_columns(df,"Nation", "Slug Nation")
```

```python
df.to_csv("./outputs/us_population.csv")
```

```python
response = requests.get("https://datausa.io/api/data?drilldowns=State&measures=Population")
state_data = response.json()["data"]
len(state_data)
```
